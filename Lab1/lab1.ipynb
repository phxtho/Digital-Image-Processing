Write a function, s = scanLine(f, I, loc), where f is a grayscale image, I is an integer, and loc is a character
string with value ‘row’ and ‘col’ indicating that I is either a row or column number in image f. The output,
s, is a vector containing the pixel values along the specified row or column in f.

```python
def scanLine(f, I=0, loc="row"):
    if loc=="row":
        return f[I,:]
    elif loc == "col":
        return f[:,I]
```

Get a horizontal intensity scan line in the middle of the image, and plot the scan line.

```python
s = scanLine(girl_image, x//2, "row")
```

```python
plt.plot(s)
plt.xlabel("pixel column")
plt.ylabel("intensity")
```

3b Scale the intensities of the image to the full [0, K] range 

fm = f − min(f);
fs = K[ fm/max(fm)]

```python
def contrastStretch(img, K):
    fmin = img.min()
    fmax = img.max()
    return np.array(list(map(lambda x : K * ((x - fmin)/(fmax - fmin)), img  )))
```

3c. Display the original image, the resulting images using K = 1, and K = 255 in (2). Is there any difference
among these 3 images?

```python
K1 = contrastStretch(spillway_img, 1)
plt.gray()
plt.imshow(K1)
```

```python
K255 = contrastStretch(spillway_img, 255)
plt.imshow(K255)
```
